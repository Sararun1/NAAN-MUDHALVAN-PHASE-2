```python
#Import required modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the dataset
df = pd.read_csv('country_vaccinations.csv')
```

```python
#Data Preprocessing

#fill missing values in numeric columns with 0
df.fillna(0, inplace=True)
```

```python
df.head()
```

```python
#Data Preprocessing

vaccine_columns = df['vaccines'].str.split(', ', expand=True)
df = pd.concat([df, vaccine_columns], axis=1)


# Rename the new columns for clarity
df.rename(columns={0: 'Vaccine_1', 1: 'Vaccine_2', 2: 'Vaccine_3', 3: 'Vaccine_4'}, inplace=True)
```

```python
df.head()
```

```python
# Drop the original "vaccines" column
df.drop('vaccines', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Identify and remove columns with no values (all zeros)
df = df.loc[:, (df != 0).any(axis=0)]
```

```python
df
```

```python
#convert the "date" column to a datetime format
df['date'] = pd.to_datetime(df['date'])
df
```

```python
#Exploratory Data Analysis

print(df.info())
```

```python
#Exploratory Data Analysis

print(df.describe())
```

```python
#Statistical Analysis

mean_total_vaccinations = df['total_vaccinations'].mean()
median_total_vaccinations = df['total_vaccinations'].median()
std_total_vaccinations = df['total_vaccinations'].std()

print(f"Mean Total Vaccinations: {mean_total_vaccinations}")
print(f"Median Total Vaccinations: {median_total_vaccinations}")
print(f"Standard Deviation Total Vaccinations: {std_total_vaccinations}")
```

```python
#Line plot to visualize the total vaccinations over time
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='date', y='total_vaccinations')
plt.title('Total Vaccinations Over Time')
plt.xlabel('Date')
plt.ylabel('Total Vaccinations')
plt.xticks(rotation=45)
plt.show()
```

```python
#Bar plot to show the most commonly used vaccines
vaccine_counts = df[['Vaccine_1', 'Vaccine_2', 'Vaccine_3', 'Vaccine_4']].stack().value_counts()
plt.figure(figsize=(10, 6))
vaccine_counts.plot(kind='bar')
plt.title('Most Commonly Used Vaccines')
plt.xlabel('Vaccine')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
